```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("2. Brain_bodyweight dataset.csv")
df.head()
```

```python
x = df['body'].values.reshape(-1,1)
y = df['brain'].values.reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 42, test_size = 0.3)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mse, mae = mean_squared_error(y_test, y_pred), mean_absolute_error(y_test, y_pred)
```

```python
print(mse, mae)
```

```python
accuracy_score = 1 - (mse/y_test.var())
accuracy_score
```

```python
plt.figure(figsize = (7,3))
plt.scatter(x_test, y_test, color = 'blue')
plt.plot(x_test, y_pred, color = 'red')
```
