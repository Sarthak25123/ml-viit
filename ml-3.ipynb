```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("3. USA_Housing.csv")
df.head()
```

```python
df.drop('Address', axis = 1, inplace = True)
```

```python
df.head()
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3)
```

```python
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
```

```python
linear.fit(x_train, y_train)
```

```python
x_train
```

```python
y_train
```

```python
predictions = linear.predict(x_test)
plt.scatter(predictions, y_test, color = 'yellow')
plt.
```

```python
plt.scatter(range(len(y_test)), y_test, color='blue', label='y_test')
plt.scatter(range(len(predictions)), predictions, color='red', label='predictions')
plt.xlabel("Prediction No.")
plt.ylabel("Price")
plt.legend()
```
